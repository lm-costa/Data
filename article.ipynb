```python
"""Importing the api class from statsmodels"""
import statsmodels.formula.api as sm
import pandas as pd
from sklearn import linear_model
import statsmodels.api as sm
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
```

```python
"x é a variavel que irá conter todos os regressores desejados"
arquivo=pd.ExcelFile('temporal.xlsx')
dados2=pd.read_excel(arquivo,'DATA')
y = dados2['XCO2']
x1 = dados2[['SIF', 'ET']]
```

```python
x1 = sm.add_constant(x1) # adding a constant
 
model = sm.OLS(y, x1).fit()
predictions = model.predict(x1) 
predictions.to_csv('model_predction.csv', index = True, header= True)
```

```python
print_model = model.summary()
print(print_model)

plt.rc('figure', figsize=(12, 7))
#plt.text(0.01, 0.05, str(model.summary()), {'fontsize': 12}) old approach
plt.text(0.01, 0.05, str(print_model), {'fontsize': 10}, fontproperties = 'monospace') # approach improved by OP -> monospace!
plt.axis('off')
plt.tight_layout()
#plt.savefig('model_summary.png', dpi = 1000)
```
